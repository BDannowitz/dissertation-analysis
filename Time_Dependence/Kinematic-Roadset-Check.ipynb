### Objectives
1. Get target dimuons from all available roadsets
2. Pick two roadsets to compare
3. Plot the normalized distributions of a single kinematic for both roadsets on the same plot
4. On top of that, plot the residuals
5. Compute $\chi^2/d.o.f.$ with hypothesis that distributions match
6. Repeat for each (1) pair of roadsets and (2) target for each (3) distribution
