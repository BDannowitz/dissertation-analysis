```python
xT_dict = {}
live_p_dict = {}
for target in target_dict:
    
    # xT values by target
    xT_dict[target] = get_xT(server, analysis_schema, analysis_table, target_dict[target]['targpos'])
    
    live_p_dict[target] = 0.0
    for merged_schema in merged_schema_list:
        # Live proton count (in units of 10^16 live protons)
        live_p_dict[target] += get_live_p_1016(server, analysis_schema, merged_schema,
                                               target_dict[target]['targpos'], bad_spill_set)
    
    print target, ': N =', len(xT_dict[target]), ', live_p:', round(live_p_dict[target],3), 'x 10^16'
```

```python
# 3. Get histogram counts and error for each target

n_bins = 20
max_x = 1.0
min_x = 0.0
bin_width = (float(max_x) - float(min_x)) / float(n_bins)
print 'Bin Width: ' + str(bin_width)

hist_dict = {}
for target in xT_dict:
    ucounts, centers = get_hist_vals(xT_dict[target], live_p_dict[target], n_bins, min_x, max_x)
    hist_dict[target] = {'ucounts': ucounts, 'bin_centers': centers}
```

```python
# Here, I find the lognormal fit for each *unbinned xT distribution*
# These PDF's are normalized to 1, so I also calculate the scale factor
#     for scaling them up the the histogram's size.

pdf_dict = {}
for target in target_dict:
    count_sum = len(xT_dict[target])
    pdf_scale = bin_width * count_sum / live_p_dict[target]
    s, loc, scale = sp.stats.lognorm.fit(xT_dict[target])
    rv = sp.stats.lognorm(s=s, scale=scale, loc=loc)
    pdf_dict[target] = {'pdf': rv, 'pdf_scale': pdf_scale}
```

```python
x = np.linspace(0, n_bins-1, n_bins)
x1 = np.linspace(0,1,1000)
```

```python
pylab.rcParams['figure.figsize'] = 8, 24

fig1 = plt.figure()
gs = gridspec.GridSpec(11, 2)
gs.update(hspace=0.7)
ax0 = fig1.add_subplot(gs[0:3, :])
ax0.set_title('# Bins = ' + str(n_bins) + ', Bin Width = ' + str(round(bin_width,3)), size=12, y=1.02)
ax1 = fig1.add_subplot(gs[3:5, :])
ax2 = fig1.add_subplot(gs[5:7, :])
ax3 = fig1.add_subplot(gs[7:9, :])
ax4 = fig1.add_subplot(gs[9:11, :])

color_dict = {'C': "#9b59b6", 'Fe': "#3498db", 'W': "#95a5a6"}

for target in ('C', 'Fe', 'W'):
    
    ax0.plot(x1, 
             pdf_dict[target]['pdf'].pdf(x1)*pdf_dict[target]['pdf_scale'],
             lw=1, color=color_dict[target], label=target + ' lognormal fit')
    ax0.errorbar(hist_dict[target]['bin_centers'], 
                 unp.nominal_values(hist_dict[target]['ucounts']),
                 yerr=unp.std_devs(hist_dict[target]['ucounts']),
                 fmt='o', color=color_dict[target], label=target + ' (N = ' + str(len(xT_dict[target])) + ')' )
    
    diff = np.subtract(hist_dict[target]['bin_centers'],hist_dict['LD2']['bin_centers'])
    ax1.plot(x, diff, lw=2, label=r'$x_T($' + target + r'$)-x_T(LD2)$')
    
    diff2 = np.subtract(pdf_dict[target]['pdf'].pdf(np.add(hist_dict[target]['bin_centers'],
                                                           diff))*pdf_dict[target]['pdf_scale'],
                        pdf_dict[target]['pdf'].pdf(hist_dict[target]['bin_centers'])*pdf_dict[target]['pdf_scale'])
    ax2.plot(x,diff2, lw=2, label=r'$Yield_{' + target + r'}(x_T + \Delta x_T)-Yield_{' + target + r'}(x_T)$')
    
    ratio = np.divide(diff2,pdf_dict[target]['pdf'].pdf(hist_dict[target]['bin_centers'])*pdf_dict[target]['pdf_scale'])
    ax3.plot(x, ratio, lw=2, label=target)
    
    ratio2 = np.divide(diff2,unp.std_devs(hist_dict[target]['ucounts']))
    ax4.plot(x, ratio2, lw=2, label=target)

ax0.set_xlabel(r'$x_T$', fontsize=14)
ax0.set_ylabel(r'Yields per $10^{16}$ live proton', fontsize=14)    
ax0.legend(loc='best')
    
ax1.set_xlabel(r'$bin#$', fontsize=14)
ax1.set_ylabel(r'$\Delta x_T$', fontsize=14)    
ax1.legend(loc='best')

ax2.set_xlabel(r'$bin#$', fontsize=14)
ax2.set_ylabel(r'$\Delta Counts$', fontsize=14)    
ax2.legend(loc='best')

ax3.set_xlabel(r'$bin#$', fontsize=14)
ax3.set_ylabel(r'$\frac{\Delta Counts}{Counts}$', fontsize=14)    
ax3.legend(loc='best')

ax4.set_xlabel(r'$bin#$', fontsize=14)
ax4.set_ylabel(r'$\frac{\Delta Counts}{\sigma(Counts)}$', fontsize=14) 
ax4.set_xlim(0,n_bins)
ax4.legend(loc='best')
```

```python
def mylognorm(x, factor, scale, mu, sigma):
    a = 1.0/((x/scale) * sigma * np.sqrt(2.0 * np.pi))
    b = - ((np.log(x/scale) - mu) ** 2) / (2 * sigma ** 2)
    p = factor * a * np.exp(b)
    return p

pinit = [25, 0.18, 0.05, 0.1]
bins = np.array(hist_dict['C']['bin_centers'])
bindata = unp.nominal_values(hist_dict['C']['ucounts'])
sig = [x if x>0 else 3 for x in unp.std_devs(hist_dict['C']['ucounts'])]
xdata = pdf_dict['C']['pdf'].pdf(x1)*pdf_dict['C']['pdf_scale']
```

```python
popt, pcov = sp.optimize.curve_fit(mylognorm, bins, bindata, p0=pinit)
```
